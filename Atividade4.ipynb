```python
import pandas as pd
from sklearn import datasets
from matplotlib import pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/sanchesprogramer/Inteligencia-Artificial/refs/heads/main/Pasta1.csv", sep = ";")
```

```python
dataset.head()
```

```python
dataset = dataset[~dataset.isin(['?', np.nan]).any(axis=1)]
dataset
```

```python
dataset.shape
```

```python
categorical_cols = dataset.select_dtypes(include=['object']).columns.tolist()
print(categorical_cols)
df_one_hot = pd.get_dummies(dataset, columns=categorical_cols, drop_first=True)
df_one_hot
```

```python
scaler = MinMaxScaler()
numeric_cols = df_one_hot.select_dtypes(include=['int64', 'float64']).columns.tolist()
df_one_hot[numeric_cols] = scaler.fit_transform(df_one_hot[numeric_cols])
df_one_hot.columns
```

```python
#PreparaÃ§Ã£o
atributos = df_one_hot.drop('aspiration_turbo', axis=1)
variavel_alvo = df_one_hot['aspiration_turbo']

atributos_encoded = pd.get_dummies(atributos, drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(atributos_encoded, variavel_alvo, test_size=0.3, random_state=42, stratify=variavel_alvo)
```

```python
#RegressÃ£o LogÃ­stica
logistic_model = LogisticRegression(max_iter=200)
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
acuracia_regressao = accuracy_score(y_test, y_pred)
print("AcurÃ¡cia:", acuracia_regressao)
```

```python
#k-Vizinhos Mais PrÃ³ximos
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
acuracia_kvizinhos = accuracy_score(y_test, y_pred)
print("AcurÃ¡cia:", acuracia_kvizinhos)
```

```python
#Decision Tree
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)
y_pred = tree_model.predict(X_test)
acuracia_decision_tree = accuracy_score(y_test, y_pred)
print("AcurÃ¡cia:", acuracia_decision_tree)
```

```python
#Floresta AleatÃ³ria
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)  # 100 Ã¡rvores
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
acuracia_floresta_aleatoria = accuracy_score(y_test, y_pred)
print("AcurÃ¡cia:", acuracia_floresta_aleatoria)
```

```python
#Gradient Boosting
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=3, random_state=42)
gb_model.fit(X_train, y_train)
y_pred = gb_model.predict(X_test)
acuracia_gradient_boosting = accuracy_score(y_test, y_pred)
print("AcurÃ¡cia:", acuracia_gradient_boosting)
```

```python
#ValidaÃ§Ã£o cruzada utilizando GridSearchCV
def train_and_evaluate_model(model, param_grid, X_train, y_train, X_test, y_test):
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=5, n_jobs=-1, verbose=2)
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)

    print("Melhores hiperparÃ¢metros:", grid_search.best_params_)

# RegressÃ£o LogÃ­stica
logistic_param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}
train_and_evaluate_model(LogisticRegression(max_iter=200), logistic_param_grid, X_train, y_train, X_test, y_test)

# k-Vizinhos Mais PrÃ³ximos
knn_param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance']
}
train_and_evaluate_model(KNeighborsClassifier(), knn_param_grid, X_train, y_train, X_test, y_test)

# Ãrvore de DecisÃ£o
tree_param_grid = {
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10]
}
train_and_evaluate_model(DecisionTreeClassifier(), tree_param_grid, X_train, y_train, X_test, y_test)

# Floresta AleatÃ³ria
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
train_and_evaluate_model(RandomForestClassifier(), rf_param_grid, X_train, y_train, X_test, y_test)

# Gradient Boosting
gb_param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}
train_and_evaluate_model(GradientBoostingClassifier(), gb_param_grid, X_train, y_train, X_test, y_test)
```

```python
#RegressÃ£o LogÃ­stica classification_report
logistic_model = LogisticRegression(max_iter=200)
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print("RegressÃ£o LogÃ­stica:")
print(classification_report(y_test, y_pred))
```

```python
#Floresta AleatÃ³ria classification_report
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

print("Floresta AleatÃ³ria:")
print(classification_report(y_test, y_pred))
```
